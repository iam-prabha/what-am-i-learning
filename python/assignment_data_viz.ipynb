```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('public_holidays_dataset.csv')
```

```python
df.info()
```

```python
df
```

```python
plt.plot(df['Date'],df['Month'], color='green', marker='*', markersize='10')
plt.grid(True)
plt.title('public holidays')
plt.xlabel('Date')
plt.ylabel('Month')
plt.show()
```

# pie chart

```python
df['total_holiday'] = df['Date'].sum()
```

```python
plt.figure(figsize=(15,10))
plt.pie(df['total_holiday'], labels=df['Name'],shadow=True)
```
