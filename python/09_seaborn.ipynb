# Library

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

# Data

```python
df = pd.read_csv('public_holidays_dataset.csv')
```

```python
df
```

# Line Plot

```python
sns.lineplot(data=df, x='Month', y='Name', hue='Day')
```

# Bar Plot

```python
mont_wise_groupby = df.groupby('Month').count()
```

```python
mont_wise_groupby
```

```python
sns.barplot(data=mont_wise_groupby, x='Month', y='Date')
```

```python
plt.figure(figsize=(20,15))
sns.barplot(data=df, x='Month', y='Date', hue='Name', palette='Set2')
```

# Histogram

```python
sns.histplot(data=df, x='Date', bins=3, kde=True) # kde-kernel Density Estimator 
```

# ScatterPlot

```python
sns.scatterplot(data=df, x='Date', y = 'Name', hue='Month')
```

# HeatMap

```python
month_dict = {'Jan': 1,
             'Feb' : 2,
             'Mar' : 3,
             'Apr' : 4,
             'May ' : 5,
             'Jun' : 6,
             'Jul' : 7,
             'Aug' : 8,
             'Sep ' : 9,
             'Oct' : 10,
             'Nov' : 11,
             'Dec' : 12}
df['Month_Number']  =df['Month'].map(month_dict)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```
